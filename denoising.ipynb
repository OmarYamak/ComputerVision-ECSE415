```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.measure import compare_psnr

#1 Filtering
#1.1 Denoising
#1.11 
img_path = "gaussian_noise_peppers.png"
img = cv2.imread(img_path)
image_path = "peppers.png"
originalimg = cv2.imread(image_path)

#Filtering
res1 = cv2.GaussianBlur(img, (7,7), 2) #7x7 Gaussian Filter with std=2
print ("PSNR", compare_psnr(originalimg, res1))
#BGR TO RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
res1 = cv2.cvtColor(res1, cv2.COLOR_BGR2RGB)

#Plot
plt.subplot(131), plt.imshow(img)
plt.title("Original"), plt.xticks([]), plt.yticks([])

plt.subplot(132), plt.imshow(res1)
plt.title("7x7 Gaussian"), plt.xticks([]), plt.yticks([])
```

```python
from skimage.measure import compare_psnr
#1.12 Box Filter
# image filtering
image_path = "gaussian_noise_peppers.png"
img = cv2.imread(image_path)

image_path = "peppers.png"
originalimg = cv2.imread(image_path)

# create a box filter, kernel is type of filter 
kernel = np.ones((7,7), np.float32)/(7*7)

# filter the image
res1 = cv2.filter2D(img, -1, kernel)
print ("PSNR Value:", compare_psnr(originalimg, res1))
# convert from BGR to RGB 
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
res1 = cv2.cvtColor(res1, cv2.COLOR_BGR2RGB)


# plot
plt.subplot(131), plt.imshow(img)
plt.title("Original"), plt.xticks([]), plt.yticks([])

plt.subplot(132), plt.imshow(res1)
plt.title("Box filter"), plt.xticks([]), plt.yticks([])
```

1.13: 
As shown above, the PSNR values are as follow: 
Gaussian Filter: 29.42329662137037
Box Filter: 28.278396476904042
Rounded to 3 sig figs, 29.4, 28.3 These values are very similar. 
This shows that the Gaussian filter gives only a slightly superior result which could be considered neglible.

```python
#1.15
image_path = "saltandpepper_noise_peppers.png"
img = cv2.imread(image_path)

image_path = "peppers.png"
originalimg = cv2.imread(image_path)

# filter image using 7x7 median filter
res = cv2.medianBlur(img, 7)
print ("PSNR Value:", compare_psnr(res, originalimg))

# convert from BGR to RGB for displaying
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
res = cv2.cvtColor(res, cv2.COLOR_BGR2RGB)

# plot
plt.subplot(121), plt.imshow(img)
plt.title("Image"), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(res)
plt.title("Median"), plt.xticks([]), plt.yticks([])
plt.show()
```

1.16: 
As shown above, the PSNR values are as follow: 
Gaussian Filter: 28.16493770977665
Median Filter: 31.871009920638333

This shows that the Median filter gives the superior result as it has a larger signal to noise ratio
