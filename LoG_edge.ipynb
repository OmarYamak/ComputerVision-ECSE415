```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np

#2.2 Laplacian of Gaussian 

# image filtering
image_path = "woman.png"
img = cv2.imread(image_path)

# filter image using 5x5 Gaussian, the std is calculated automatically
res1 = cv2.GaussianBlur(img, (5,5), 0)

#convert to grayscale before applying laplacian 
res1= cv2.cvtColor(res1, cv2.COLOR_BGR2GRAY)

# apply 5x5 Laplacian filter to the intensity channel
lap = cv2.Laplacian(res1, cv2.CV_32F, ksize=5)

# display images
plt.figure(figsize=(20, 20))
plt.subplot(211), plt.imshow((lap), cmap="gray")
plt.title("Image"), plt.xticks([]), plt.yticks([])
```
